# Importations

```python
import numpy as np
import csv

import pandas as pd
```

# Import data

```python
data = []
with open('dtu data_final.csv',"r") as infile:
    reader=csv.DictReader(infile, delimiter=';')
    for row in reader:
        data.append(row)

for d in data:
    d['ADHD'] = float(d['ADHD'].replace(',',"."))
    d['Diabetes'] = float(d['Diabetes'].replace(',',"."))
    d['OTC'] = float(d['OTC'].replace(',',"."))
    d['Branded'] = float(d['Branded'].replace(',',"."))
    d['Rygestop'] = float(d['Rygestop'].replace(',',"."))
    d['Astma'] = float(d['Astma'].replace(',',"."))
    d['RX'] = float(d['RX'].replace(',',"."))
    d['Modul'] = float(d['Modul'].replace(',',"."))
```

## Keep numbers

```python
exclude = ["Habitat", "Region", "Chain", "OTC", "RX"]
categories = [a for a in categories if a not in exclude]
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.palettes import brewer

df = pd.DataFrame(data)
df.drop(exclude, axis=1, inplace=True)
df = df.set_index(['Modul'])
```

# Stacked plot representing 

```python
categories = ["ADHD", "Astma", "Rygestop", "Diabetes", "Branded"]

x=[]
for d in data:
    x.append(d['Modul'])
    
somme = df['ADHD']+df['Astma']+df['Diabetes']+df['Rygestop']+df["Branded"]
somme = pd.DataFrame(somme)

for i in range(len(df.values)):
    df.values[i,:] = df.values[i,:]/somme.values[:,0][i]

def stacked(df, categories):
    areas = dict()
    last = np.zeros(len(df[categories[0]]))
    for cat in categories:
        next = last + df[cat]
        areas[cat] = np.hstack((last[::-1], next))
        last = next
    return areas

areas = stacked(df, categories)

colors = brewer["Spectral"][len(areas)]

x2 = np.hstack((x[::-1], x))

p = figure()
p.grid.minor_grid_line_color = '#eeeeee'

p.patches([x2] * len(areas), [areas[cat] for cat in categories],
          color=colors, alpha=0.8, line_color=None)

output_file("brewer.html", title="cashflow")

show(p)
```
